```python
import pandas as pd
import numpy as np
```

```python
m=pd.read_csv('Z:\\Z 5\\ML\\text_classification.csv')
m
```

```python
train=m[:10]
test=m[10:]
```

```python
l=[]
for i in train['Text']:
    l.extend(i.split(' '))
l=set(l)
```

```python
df=pd.DataFrame([[0]*len(l)]*len(train[train['Type']=='pos']),columns=l)
df
```

```python
c=0
# FOR POSITIVE PROBABILITY


for i in train[train['Type']=='pos'].iterrows():
    if (i[1][1]=='pos'):
        temp=i[1][0].split(' ')
        for j in temp:
            if j in df:
                df[j][c]+=1
    c+=1
prob_pos={}
total=sum(df.sum(axis=1))+len(df.columns)
for i in df.columns:
    prob_pos[i]=(df[i].sum()+1)/total
prob_pos
```

```python
# for negative probability

c=0
df_neg=pd.DataFrame([[0]*len(l)]*len(train[train['Type']=='neg']),columns=l)
for i in train[train['Type']=='neg'].iterrows():
    if (i[1][1]=='neg'):
        temp=i[1][0].split(' ')
        for j in temp:
            if j in df:
                df_neg[j][c]+=1
    c+=1
df_neg
prob_neg={}
total=sum(df_neg.sum(axis=1))+len(df_neg.columns)
for i in df_neg.columns:
    prob_neg[i]=(df_neg[i].sum()+1)/total
prob_neg
```

```python
res=[]
for i in test.iterrows():
    pos_prob=len(train[train['Type']=='pos'])/len(train)
    neg_prob=len(train[train['Type']=='neg'])/len(train)
    for j in i[1][0].split(' '):
        if j in df:
            pos_prob*=prob_pos[j]
            neg_prob*=prob_neg[j]
    print('Positive:',pos_prob,"\nNegative",neg_prob)
    if (pos_prob>neg_prob):
        print('Positive')
        res.append('pos')
    else:
        print('Negative')
        res.append('neg')
```

# Accuracy Precision Recall

```python
tp,tn,fp,fn=0,0,0,0
k=0
tp=min(res.count('pos'),len(test[test['Type']=='pos']))
tn=min(res.count('neg'),len(test[test['Type']=='neg']))
for i in test.iterrows():
    if i[1][1]=='pos' and res[k]=='neg':
        fn+=1
    elif i[1][1]=='neg' and res[k]=='pos':
        fp+=1
    k+=1
accuracy=(tp+tn)/(tp+tn+fp+fn)
precision=tp/(tp+fp)
recall=tp/(tp+fn)
print(tp," ",tn,' ',fp,' ',fn)
print("Accuracy: ",accuracy)
print('Precission: ',precision)
print('Recall: ',recall)
```
